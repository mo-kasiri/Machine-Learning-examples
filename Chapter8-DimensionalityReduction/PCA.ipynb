```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
n_points = 100

# Create a 3D cloud: x, y are correlated, z adds some noise
x = np.random.rand(n_points) * 10
y = 2 * x + np.random.randn(n_points) * 2
z = 0.5 * x + np.random.randn(n_points) * 1.5
```

```python
X = np.vstack((x, y, z)).T 
X.shape
```

```python
# Centering the data
X_centered = X - X.mean(axis=0)
X_centered.mean()
```

```python
# SVD
U, S, Vt = np.linalg.svd(X_centered)
# Principal components
PCs = Vt.T
```

```python
X_2D = X_centered.dot(PCs[:, :2])  # shape (100, 2)
```

```python
X_2D
```

```python
fig = plt.figure(figsize=(12, 5))

# 3D scatter
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c='blue', alpha=0.6)
ax1.set_title("Original 3D Data")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")

# 2D projection
ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(X_2D[:, 0], X_2D[:, 1], c='red', alpha=0.6)
ax2.set_title("Projected onto 2D (first 2 PCs)")
ax2.set_xlabel("PC1")
ax2.set_ylabel("PC2")

plt.tight_layout()
plt.show()
```
