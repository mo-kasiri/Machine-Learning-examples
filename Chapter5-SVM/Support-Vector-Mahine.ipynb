```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
iris.target_names
```

```python
data = iris["data"][:, (2,3)]
target = iris["target"]
```

```python
target
```

```python
plt.plot(data[:, 0][iris["target"]==0], data[:, 1][iris["target"]==0], "b.", label="Iris setosa")
plt.plot(data[:, 0][iris["target"]==1], data[:, 1][iris["target"]==1], "y.", label="Iris versicolor")
plt.plot(data[:, 0][iris["target"]==2], data[:, 1][iris["target"]==2], "r.", label="Iris virginica")
plt.axis([0, 8, 0, 3])
plt.grid(True, which='both')
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.legend(loc="upper left", fontsize=14)
plt.show()
```

```python
X = iris["data"][:, (2,3)] # petal length, petal width 
y = (iris["target"] == 2).astype(np.float64) # Iris virginica
```

```python
svm_clf = Pipeline([ ("scalar", StandardScaler()), ("linear_svc", SVC(kernel="linear", C=1)) ])
svm_clf.fit(X,y)
```

```python
svm_clf.predict([[5.5, 1.5]])
```
